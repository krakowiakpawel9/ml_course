```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
sns.set()
np.random.seed(42)
sklearn.__version__
```

```
data = np.arange(20)
target = 2 * data + 10 + 3 * np.random.randn(20)

data = data.reshape(-1, 1)
```

```
plt.title('Outliers')
plt.scatter(data, target, label='cecha x')
plt.legend()
plt.xlabel('cecha x')
plt.ylabel('target')
plt.show()
```

```
from sklearn.linear_model import LinearRegression

regressor_1 = LinearRegression()
regressor_1.fit(data, target)

from sklearn.metrics import mean_absolute_error

mae_1 = mean_absolute_error(target, regressor_1.predict(data))
print(f'MAE_1 = {mae_1:.4f}')
```

```
plt.title(f'MAE_1 = {mae_1:.4f}')
plt.scatter(data, target, label='cecha x')
plt.plot(data, regressor_1.intercept_ + regressor_1.coef_[0] * data,
         label='regressor_1', c='red')
plt.legend()
plt.xlabel('cecha x')
plt.ylabel('target')
plt.show()
```

```
# target[18] = 105. 
target[19] = 100.5

regressor_2 = LinearRegression()
regressor_2.fit(data, target)
mae_2 = mean_absolute_error(target, regressor_2.predict(data))
print(f'MAE = {mae_2:.4f}')
```

```
plt.title(f'MAE_1 = {mae_1:.4f} MAE_2 = {mae_2:.4f}')
plt.scatter(data, target, label='cecha x')
plt.plot(data, regressor_1.intercept_ + regressor_1.coef_[0] * data,
         label='regressor_1', c='red')
plt.plot(data, regressor_2.intercept_ + regressor_2.coef_[0] * data,
         label='regressor_2', c='blue')
plt.legend()
plt.xlabel('cecha x')
plt.ylabel('target')
plt.show()
```

```
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=200.0)
ridge.fit(data, target)

mae_3 = mean_absolute_error(target, ridge.predict(data))
print(f'MAE = {mae_3:.4f}')
```

```
plt.title(f'MAE_1 = {mae_1:.4f} MAE_2 = {mae_2:.4f} MAE_3 = {mae_3:.4f}')
plt.scatter(data, target, label='cecha x')
plt.plot(data, regressor_1.intercept_ + regressor_1.coef_[0] * data,
         label='regressor_1', c='red')
plt.plot(data, regressor_2.intercept_ + regressor_2.coef_[0] * data,
         label='regressor_2', c='blue')
plt.plot(data, ridge.intercept_ + ridge.coef_[0] * data,
         label='ridge', c='green')
plt.legend()
plt.xlabel('cecha x')
plt.ylabel('target')
plt.show()
```
