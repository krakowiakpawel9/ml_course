````
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
sns.set(font_scale=1.3)

sklearn.__version__
````

````
from sklearn.datasets import make_regression

data, target = make_regression(n_samples=100, n_features=1, n_targets=1, noise=30.0)
print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
````

````
data[:10]
````

````
target[:10]
````

````
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data, target, label='cecha x')
plt.legend()
plt.plot()
````

````
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

# metoda fit() dopasowuje model liniowy
regressor.fit(data, target)
````

````
# metoda predict() dokonuje predykcji na podstawie modelu
y_pred = regressor.predict(data)
y_pred
````

````
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data, target, label='cecha x')
plt.plot(data, y_pred, color='red', label='model')
plt.legend()
plt.show()
````

Współczynnik $R^2$ jest zdefiniowany jako $1 - \frac{u}{v}$, gdzie $u$ jest określone wzorem `((y_true - y_pred) ** 2).sum()` oraz $v$ jako `((y_true - y_true.mean()) ** 2).sum()`. 
Czyli 
```
1 - (((y_true - y_pred) ** 2).sum()) / (((y_true - y_true.mean()) ** 2).sum())
```
Najlepszą możliwą wartością jest 1.0 oraz może przyjmować wartość ujemną. Model, który przewiduje zawsze dla każdej próbki wartość oczekiwaną zmiennej docelowej ma współczynnik $R^2$ równy 0.

````
# metoda score() w przypadku regresji liniowej zwraca współczynnik determinacji R2
regressor.score(data, target)
````

````
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data, target, label='cecha x')
plt.plot(data, regressor.intercept_ + regressor.coef_[0] * data, color='red', label='model')
plt.legend()
plt.show()
````

Regresja z podziałem na zbiór treningowy oraz testowy

````
data, target = make_regression(n_samples=1000, n_features=1, n_targets=1, noise=25.0)
print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
````

````
from sklearn.model_selection import train_test_split

data_train, data_test, target_train, target_test = train_test_split(data, target, test_size=0.25)

print(f'data_train shape: {data_train.shape}')
print(f'data_test shape: {data_test.shape}')
print(f'target_train shape: {target_train.shape}')
print(f'target_train shape: {target_train.shape}')
````

````
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa train vs. test')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data_train, target_train, label='zbiór treningowy', color='blue', alpha=0.5)
plt.scatter(data_test, target_test, label='zbiór testowy', color='green', alpha=0.5)
plt.legend()
plt.plot()
````

````
regressor = LinearRegression()
regressor.fit(data_train, target_train)
````

````
regressor.score(data_train, target_train)
````

````
regressor.score(data_test, target_test)
````

````
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa: zbior treningowy')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data_train, target_train, label='zbiór treningowy', color='blue', alpha=0.5)
plt.plot(data_train, regressor.intercept_ + regressor.coef_[0] * data_train, color='red')
plt.legend()
plt.plot()
````

````
plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa: zbior testowy')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data_test, target_test, label='zbiór testowy', color='green', alpha=0.5)
plt.plot(data_test, regressor.intercept_ + regressor.coef_[0] * data_test, color='red')
plt.legend()
plt.plot()
````

````
y_pred = regressor.predict(data_test)

predictions = pd.DataFrame(data={'y_true': target_test, 'y_pred': y_pred})
predictions.head()
````

````
predictions['error'] = predictions['y_true'] - predictions['y_pred']
predictions.head()
````
