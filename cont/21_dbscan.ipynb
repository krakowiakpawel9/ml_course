```
import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import make_blobs
import plotly.express as px
sklearn.__version__
```

```
data, _ = make_blobs(n_samples=1000, centers=3, cluster_std=1.2, center_box=(-8.0, 8.0), random_state=42)
df = pd.DataFrame(data, columns=['x1', 'x2'])
px.scatter(df, 'x1', 'x2', width=700, height=500, title='Klasteryzacja')
```

```
from sklearn.cluster import DBSCAN

cluster = DBSCAN(eps=0.8, min_samples=5)
cluster.fit(data)
```

```
cluster.labels_[:10]
```

```
df['cluster'] = cluster.labels_
px.scatter(df, 'x1', 'x2', 'cluster', width=700, height=500, title='Klasteryzacja')
```
