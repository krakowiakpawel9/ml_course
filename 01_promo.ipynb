### <a name='0'></a> Import bibliotek

```
import pandas as pd
import pandas_datareader.data as web

pd.set_option('precision', 2)
pd.set_option('display.float_format', '{:.2f}'.format)
```

### <a name='1'></a> Załadowanie danych

```
df = web.DataReader(name='GOOGL', data_source='stooq')
df
```

### <a name='2'></a> Eksploracja danych

```
df.describe()
```

```
df = df.reset_index()
df.head()
```

```
df_2018 = df.query("Date > '2018-01-01'")
df_2018
```

### <a name='3'></a> Eksport do pliku .xlsx

```
df_2018.to_excel('google_2018.xlsx', sheet_name='GOOGL', index=False)
```

### <a name='4'></a> Wczytanie pliku .xlsx

```
df_new = pd.read_excel('google_2018.xlsx', sheet_name='GOOGL')
df_new
```

### <a name='5'></a> Wizualizacje

```
import plotly.express as px

px.line(df_new, x='Date', y='Close', title='Notowania spółki Google (GOOGL)', 
        width=900, height=500, log_y=True, color_discrete_sequence=['#03fcc2'],
        template='plotly_dark')
```

```
px.area(df_new, x='Date', y='Close', title='Notowania spółki Google (GOOGL)', 
        width=900, height=500, log_y=True, color_discrete_sequence=['#c50ff7'],
        template='plotly_dark')
```

```
import plotly.graph_objects as go

fig = go.Figure(data=[
    go.Candlestick(
        x=df_new['Date'],
        open=df_new['Open'],
        high=df_new['High'],
        low=df_new['Low'],
        close=df_new['Close']
    )
])
fig.update_layout(title='Notowania spółki Google (GOOGL)',
                  width=900, height=500, template='plotly_dark')
fig.show()
```
