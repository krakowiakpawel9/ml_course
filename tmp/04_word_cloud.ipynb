### <a name='0'></a> Import bibliotek

````
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
````

````
text = """Apple's chief executive said the company would open its first physical stores in India in 2021 and a online outlet later this year. 
Apple had to seek special approval from the Indian government to open a store without a local partner. 
The announcement was made at the company's annual shareholders' meeting. 
Investors at the meeting also voted on a proposal that the firm should alter how it responds when governments ask it to remove apps from its marketplace. 
Though the measure wasn't approved, it failed by a slimmer margin then similar proposals in the past. 
Apple's move into India, the second largest smartphone market in the world, has been expected for some time, but the announcement of a date was new.
In 2018 India changed the laws that prevented foreign brands from opening single-brand stores in the country.
Nevertheless said India had wanted Apple to open its store with a local partner."""
````

````
mask = np.zeros((400, 400), dtype='uint8')
mask
````

````
wc = WordCloud(mask=mask, random_state=41)
wc.generate(text)

plt.figure(figsize=(6, 6))
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
````

````
wc = WordCloud(background_color='white', mask=mask, random_state=41)
wc.generate(text)

plt.figure(figsize=(6, 6))
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
````

````
wc = WordCloud(background_color='yellow', mask=mask, random_state=41)
wc.generate(text)

plt.figure(figsize=(6, 6))
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
````

````
wc = WordCloud(background_color='white', mask=mask, max_font_size=52, random_state=41)
wc.generate(text)

plt.figure(figsize=(6, 6))
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
````

````
import cv2
from google.colab.patches import cv2_imshow
import imutils

img = cv2.imread('logo.png')
img.shape
````

````
img = imutils.resize(img, height=500)
cv2_imshow(img)
````

````
gray = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
````

````
mask_inv = cv2.bitwise_not(gray)
cv2_imshow(mask_inv)
````

````
mask = cv2.threshold(src=mask_inv, thresh=220, maxval=255, type=cv2.THRESH_BINARY)[1]
cv2_imshow(mask)
````

````
mask = cv2.copyMakeBorder(mask, top=10, bottom=10, left=10, right=10,
                          borderType=cv2.BORDER_CONSTANT, value=[255, 255, 255])
cv2_imshow(mask)
````

````
wc = WordCloud(background_color="white", mask=mask)
wc.generate(text)
wc.to_file('logo_wc.png')

plt.figure(figsize=(6, 6))
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
````

````
wc = WordCloud(background_color="white", mask=mask, contour_width=5, contour_color='grey')
wc.generate(text)
wc.to_file('logo_wc.png')

plt.figure(figsize=(6, 6))
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
````
