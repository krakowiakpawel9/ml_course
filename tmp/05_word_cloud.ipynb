````
from PyPDF2 import PdfFileReader

with open('playway.pdf', 'rb') as file:
    pdf = PdfFileReader(file)
    text = ''
    for page_idx in range(pdf.getNumPages()):
        page = pdf.getPage(page_idx)
        text += page.extractText()

print(text)
````

````
from wordcloud import WordCloud

mask = np.zeros((400, 400), dtype='uint8')

stopwords = ['za', 'od', 'do']

wc = WordCloud(background_color='white', mask=mask, stopwords=stopwords, random_state=41)
wc.generate(text)

plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
````

````
wc = WordCloud(background_color='white', mask=mask, stopwords=stopwords, min_font_size=8, random_state=41)
wc.generate(text)

plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(wc, interpolation='bilinear')
plt.show()
````
